```python
import pandas as pd
```

```python
train_df = pd.read_csv('test.csv')
```

```python
# Load model Preprocessing and Model
train_df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder


# initialize a LabelEncoder object
le = LabelEncoder()

# apply label encoding to the 'color' column
train_df['dialect_index'] = le.fit_transform(train_df['dialect'])

```

```python
train_df.head(20)
```

```python
import joblib

pipeline = joblib.load('/content/DL1_pipeline.pkl')
```

```python
import numpy as np
y_prob = pipeline.predict(train_df.text)
y_pred = np.argmax(y_prob, axis=-1)
```

```python
from sklearn import metrics


report = metrics.classification_report(train_df.dialect_index, y_pred)
print(report)

print("accuracy: {:0.3f}".format(metrics.accuracy_score(train_df.dialect_index, y_pred)))
```
