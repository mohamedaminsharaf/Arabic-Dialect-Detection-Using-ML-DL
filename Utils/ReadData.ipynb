```python
import pandas as pd
import sqlite3
```

```python
# Connect to the database file
conn = sqlite3.connect('dialects_database.db')

# Create a cursor object to execute SQL commands
cur = conn.cursor()

# Execute a query to retrieve the names of all tables in the database
query = """
SELECT 
    name,
    sql 
FROM sqlite_master 
WHERE type='table'
"""
cur.execute(query)

# Fetch all rows of the query result
rows = cur.fetchall()

# Loop through the rows and print the table names
for row in rows:
    print(row[0],'\n',row[1])
```

```python
# Execute a SELECT query to retrieve data from the database
query = """
SELECT * 
FROM id_text
LIMIT 5
"""
cur.execute(query)

# Fetch all rows of the query result
rows = cur.fetchall()

# Loop through the rows and print the data
for row in rows:
    print(row)
```

```python
# Execute a SELECT query to retrieve data from the database
query = """
SELECT * 
FROM id_dialect
LIMIT 5
"""
cur.execute(query)

# Fetch all rows of the query result
rows = cur.fetchall()

# Loop through the rows and print the data
for row in rows:
    print(row)
```

```python
query = """
SELECT 
    id_text.id, 
    text, 
    dialect 
FROM id_text, id_dialect
WHERE id_text.id = id_dialect.id"""

df = pd.read_sql(query,conn)
df.head()
```

```python
# Close the cursor and connection objects
cur.close()
conn.close()
```
